## 🔄 Comparación: Regresión vs Clasificación

En el tema anterior, trabajamos con **Regresión Lineal**. Ahora vamos a cambiar de enfoque con **Clasificación** (regresión logística). Es importante entender la diferencia fundamental:

### 📊 ¿Qué es Regresión?

En **regresión**, el modelo **predice un valor numérico continuo**. 

**Ejemplo del tema anterior: Predicción de Salarios**
- **Entrada**: Años de experiencia (número continuo: 5.0, 6.5, 15.3...)
- **Salida**: Salario predicho (número continuo: 45,250€, 72,300€...)
- **Pregunta que hace el modelo**: "¿Cuál es el salario probable para un empleado con X años de experiencia?"
- **Tipo de respuesta**: Un número específico

La salida es una **cantidad**, algo que se mide en una escala continua (euros, metros, kilos, etc.).

### 🎯 ¿Qué es Clasificación?

En **clasificación**, el modelo **predice la categoría o clase a la que pertenece un objeto**.

**Ejemplo que veremos ahora: Predicción de Churn (Abandono)**
- **Entrada**: Antigüedad del cliente, gasto mensual (números continuos)
- **Salida**: ¿Abandona el cliente? → **SÍ (1)** o **NO (0)**
- **Pregunta que hace el modelo**: "¿Este cliente va a abandonar el servicio o no?"
- **Tipo de respuesta**: Una categoría, no un número exacto

### 📈 Comparación Visual

| Característica | **Regresión** | **Clasificación** |
|:---|:---|:---|
| **¿Qué predice?** | Valores continuos (números reales) | Categorías discretas (clases) |
| **Ejemplos de salida** | 45,250€, 72,300€, 55,000€... | Churn=SÍ, Churn=NO |
| **Gráficamente** | Una línea o curva suave | Puntos agrupados por clase |
| **Pregunta típica** | "¿Cuánto?" | "¿Cuál?" o "¿Es...?" |
| **Ejemplo real** | Precio de casa, temperatura, ventas | Email spam/no-spam, diagnóstico médico, cliente leal/infiel |
| **Función del modelo** | 𝑦 = 𝜃₀ + 𝜃₁𝑥 (Regresión Lineal) | 𝑝 = 1/(1+e⁻ˢ) (Regresión Logística) |

### 🤔 ¿Cómo Evaluamos Cada Uno?

Por esta diferencia fundamental, **no podemos usar las mismas métricas** para regresión y clasificación:

- **En Regresión**: Medimos qué tan cerca está la predicción del valor real. Usamos **MAE, MSE, RMSE, R²**.
  - Pregunta: "¿Cuánto se equivocó el modelo en euros?"

- **En Clasificación**: No tiene sentido preguntar "cuánto se equivocó" (¿en qué unidades?). En cambio, contamos:
  - ¿Cuántas predicciones fueron correctas? (**Accuracy**)
  - ¿Cuántos casos positivos detectamos correctamente? (**Recall**)
  - ¿Qué tan seguros estamos de nuestras predicciones positivas? (**Precisión**)
  - Usamos la **Matriz de Confusión, Precision, Recall, F1-Score, AUC-ROC**.

### 💡 Ejemplo Práctico del Cambio

**Modelo de Regresión:**
```
Input: 5 años de experiencia
Output: 87,500€ ← Predicción exacta de un número
Real: 88,000€
Error: 500€ ← Podemos cuantificarlo
```

**Modelo de Clasificación:**
```
Input: Cliente con gasto mensual de 150€, antigüedad de 6 meses
Output: Churn = SÍ (1) ← Predicción de una clase
Real: Churn = SÍ (1)
Resultado: ✅ Acierto (Verdadero Positivo)

Input: Cliente con gasto mensual de 200€, antigüedad de 24 meses
Output: Churn = NO (0) ← Predicción de una clase
Real: Churn = SÍ (1)
Resultado: ❌ Error (Falso Negativo)
```

## 🤔 1. La Importancia de la Métrica Correcta <a id='1-importancia'></a>

Imagina dos escenarios:

1.  **Diagnóstico de Cáncer**: Un modelo que predice si un tumor es maligno.

- En la **recomendación de vídeos**, un **Falso Positivo** (recomendar un vídeo que no te gusta) es molesto, pero no grave. Queremos maximizar los aciertos, pero un error no tiene consecuencias serias.

La elección de la métrica de evaluación depende directamente del **objetivo de negocio** y de las **consecuencias de los diferentes tipos de errores**. No existe "la mejor métrica" para todos los casos.

## 📊 2. Métricas para Problemas de Regresión <a id='2-regresion'></a>

En regresión, no predecimos categorías, sino valores continuos. Por lo tanto, las métricas se basan en medir la **distancia** o el **error** entre el valor real (`y`) y el valor predicho (`ŷ`).

Para explicar estas métricas, usaremos un problema de regresión: predecir el precio de casas basándonos en su tamaño.

````python
# Preparación: Entrenemos un modelo de regresión para tener predicciones que evaluar
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Datos sintéticos de precios de casas
np.random.seed(42)
tamaño = np.random.randint(50, 200, 100)
precio_real = tamaño * 1500 + np.random.normal(0, 25000, 100)
df_casas = pd.DataFrame({'Tamaño': tamaño, 'Precio': precio_real})

X_casas = df_casas[['Tamaño']]
y_casas = df_casas['Precio']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_casas, y_casas, test_size=0.2, random_state=42)

reg_model = LinearRegression()
reg_model.fit(X_train_c, y_train_c)
y_pred_c = reg_model.predict(X_test_c)

print("Modelo de regresión entrenado y predicciones listas para evaluar.")
````

### 2.1 Error Absoluto Medio (MAE) <a id='2.1-mae'></a>

Es el promedio de las diferencias absolutas entre los valores reales y los predichos.

$$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

- **Interpretación**: "En promedio, nuestro modelo se equivoca en X unidades".
- **Ventaja**: Fácil de interpretar porque está en las mismas unidades que la variable objetivo.
- **Desventaja**: No penaliza más los errores grandes.

````python
# Cálculo de MAE (Mean Absolute Error - Error Absoluto Medio)
mae = mean_absolute_error(y_test_c, y_pred_c)
print(f"MAE: {mae:.2f}€")
````

### 2.2 Error Cuadrático Medio (MSE) <a id='2.2-mse'></a>

Es el promedio de los errores al cuadrado.

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

- **Ventaja**: Penaliza mucho más los errores grandes que los pequeños (debido al cuadrado). Es muy útil si los errores grandes son especialmente indeseables.
- **Desventaja**: Las unidades están al cuadrado (ej: €²), lo que lo hace difícil de interpretar.

````python
# Cálculo de MSE (Mean Squared Error - Error Cuadrático Medio)
mse = mean_squared_error(y_test_c, y_pred_c)
print(f"MSE: {mse:.2f}€²")
````

### 2.3 Raíz del Error Cuadrático Medio (RMSE) <a id='2.3-rmse'></a>

Es simplemente la raíz cuadrada del MSE. 

$$ \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} $$

- **Ventaja**: Combina lo mejor del MAE y el MSE. Penaliza los errores grandes (como el MSE) pero está en las mismas unidades que la variable objetivo (como el MAE), haciéndolo interpretable.
- **Es una de las métricas más utilizadas en regresión.**

````python
# Cálculo del RMSE (Raiz del error cuadrático medio)
rmse = np.sqrt(mse)
print(f"RMSE: {rmse:.2f}€")
````

### 2.4 Coeficiente de Determinación (R²) <a id='2.4-r2'></a>

R² (R-cuadrado) mide la **proporción de la varianza** en la variable dependiente que es predecible a partir de las variables independientes.

- **Interpretación**: "Nuestro modelo explica el X% de la variabilidad en los precios de las casas".
- **Rango**: Va de -∞ a 1.
  - **R² = 1**: El modelo explica toda la variabilidad (predicciones perfectas).
  - **R² = 0**: El modelo no explica nada de la variabilidad (es tan bueno como predecir siempre la media).
  - **R² < 0**: El modelo es peor que predecir la media.

**Cuidado**: R² tiende a aumentar si añades más variables al modelo, incluso si no son útiles. Para eso existe el **R² ajustado**, que penaliza la adición de variables innecesarias.

````python
# Cálculo del R2 (Coeficiente de Determinación)
r2 = r2_score(y_test_c, y_pred_c)
print(f"R²: {r2:.2f}")
print(f"Esto significa que nuestro modelo explica aproximadamente el {r2*100:.0f}% de la variabilidad en los precios de las casas.")
````

## 🎯 3. Métricas para Problemas de Clasificación <a id='3-clasificacion'></a>

Para explicar estas métricas, usaremos un problema de clasificación binaria: predecir si un cliente abandonará o no una compañía (Churn).

- **Clase Positiva (1)**: El cliente SÍ abandona (Churn).
- **Clase Negativa (0)**: El cliente NO abandona (No Churn).

````python
# Preparación: Entrenemos un modelo simple para tener predicciones que evaluar
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split # División de datos
from sklearn.linear_model import LogisticRegression # Modelo de regresión logística
# Métricas de clasificación
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report 

# Datos sintéticos de Churn
# Aquí generamos un conjunto de datos sintéticos para simular el problema de churn
np.random.seed(42) # Semilla para reproducibilidad
data = {
    'antiguedad': np.random.randint(1, 60, 200),
    'gasto_mensual': np.random.uniform(20, 200, 200),
    'churn': np.random.randint(0, 2, 200)
}
# Forzar un desbalanceo para que las métricas sean más interesantes
data['churn'][:140] = 0 # 70% no churn
data['churn'][140:] = 1 # 30% churn

# Crear DataFrame con los datos
df_churn = pd.DataFrame(data).sample(frac=1, random_state=42).reset_index(drop=True)

# Dividir en conjunto de entrenamiento y prueba
X = df_churn[['antiguedad', 'gasto_mensual']] # Características 
y = df_churn['churn'] # Etiqueta objetivo

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Entrenar un modelo simple de regresión logística
model = LogisticRegression(random_state=42) # Modelo de regresión logística
model.fit(X_train, y_train) # Entrenamiento del modelo
y_pred = model.predict(X_test) # Predicciones del modelo

print("Modelo entrenado y predicciones listas para evaluar.")
````

### 3.1 La Matriz de Confusión: La Base de Todo <a id='3.1-matriz'></a>

La matriz de confusión es una tabla que resume el rendimiento de un modelo de clasificación. Compara las etiquetas reales con las predicciones del modelo.

|                | **Predicción: NO (0)** | **Predicción: SÍ (1)** |
|----------------|------------------------|------------------------|
| **Real: NO (0)** | Verdadero Negativo (TN) | Falso Positivo (FP)    |
| **Real: SÍ (1)** | Falso Negativo (FN)    | Verdadero Positivo (TP) |

- **Verdadero Positivo (TP)**: El modelo predijo 'Churn' y el cliente realmente hizo 'Churn'. ¡Acierto!
- **Verdadero Negativo (TN)**: El modelo predijo 'No Churn' y el cliente realmente 'No Churn'. ¡Acierto!
- **Falso Positivo (FP) / Error Tipo I**: El modelo predijo 'Churn', pero el cliente se quedó. (Ej: Ofrecer un descuento innecesariamente).
- **Falso Negativo (FN) / Error Tipo II**: El modelo predijo 'No Churn', pero el cliente se fue. (Ej: Perder un cliente por no actuar a tiempo). **¡Este suele ser el más costoso!**

````python
import seaborn as sns # para visualización
import matplotlib.pyplot as plt # para visualización

cm = confusion_matrix(y_test, y_pred) # calcular la matriz de confusión
# Le pasamos y_test (etiquetas reales) y y_pred (etiquetas predichas por el modelo)
print("Matriz de Confusión:")
print(cm)

# Visualización de la Matriz de Confusión
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', # Añadido etiquetas para mayor claridad
            xticklabels=['Pred. No Churn', 'Pred. Churn'], 
            yticklabels=['Real No Churn', 'Real Churn'])
plt.title('Matriz de Confusión')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()
````

#### 📊 Interpretación de Tu Matriz de Confusión

En tu gráfico aparecen estos números exactos:

```
                 Pred. No Churn  |  Pred. Churn
Real No Churn        42          |      0
Real Churn           18          |      0
```

**Desglose por cuadrante:**

| Posición | Valor | Nombre | Significado |
|:---|:---|:---|:---|
| **Superior Izquierda** | **42** | TN (Verdadero Negativo) | ✅ 42 clientes que NO abandonaron → Modelo predijo bien |
| **Superior Derecha** | **0** | FP (Falso Positivo) | ✅ 0 clientes incorrectamente predichos como "Churn" |
| **Inferior Izquierda** | **18** | FN (Falso Negativo) | ❌❌ 18 clientes que SÍ abandonaron → Pero el modelo dijo "NO Churn" |
| **Inferior Derecha** | **0** | TP (Verdadero Positivo) | ❌ 0 clientes en churn detectados correctamente |

#### 🚨 Diagnóstico Crítico:

**El modelo es inútil para detectar churn:**
- ✅ **Lo hace bien**: Identifica clientes que NO se van (42 casos)
- ❌ **Lo hace mal**: Nunca detecta clientes que SÍ se van
- ❌ **Problema**: 18 clientes reales abandonaron sin que el modelo levantara ninguna alarma

**Métricas resultantes:**
- **Recall = 0%** → No detecta NINGÚN caso de churn real
- **Precisión = indefinida** → No puede calcularse (nunca predice positivos)
- **El modelo predice "NO Churn" para TODOS los clientes**

````python
# Análisis detallado de la matriz
tn, fp, fn, tp = cm.ravel()

print(f"📊 ANÁLISIS DETALLADO DE LA MATRIZ:")
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
print(f"TN (Superior Izq):        {tn} - Clientes leales bien identificados ✅")
print(f"FP (Superior Der):        {fp} - Falsos positivos (ninguno) ✅")
print(f"FN (Inferior Izq):        {fn} - CLIENTES PERDIDOS sin detectar ❌❌")
print(f"TP (Inferior Der):        {tp} - Churns detectados (ninguno) ❌")
print(f"━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")
print(f"\n📊 TOTALES:")
print(f"Clientes que NO se fueron: {tn + fp}")
print(f"Clientes que SÍ se fueron: {fn + tp}")
print(f"Total de predicciones: {tn + fp + fn + tp}")

if (fn + tp) > 0:
    deteccion = 100*tp/(fn+tp)
else:
    deteccion = 0
    
print(f"\n🚨 PROBLEMA CRÍTICO:")
print(f"De los {fn + tp} clientes que abandonaron: {tp} fueron detectados ({deteccion:.1f}%)")
print(f"El modelo FALLÓ en {fn} casos importantes (clientes perdidos)")
````

### 3.2 Accuracy (Exactitud) <a id='3.2-accuracy'></a>

Es la métrica más intuitiva: ¿Qué porcentaje de predicciones fueron correctas?

$$ \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}} = \frac{\text{Predicciones Correctas}}{\text{Total de Predicciones}} $$

**Cuidado**: El accuracy puede ser engañoso en **conjuntos de datos desbalanceados**. Si el 99% de tus clientes no hacen churn, un modelo que *siempre* predice "No Churn" tendrá un 99% de accuracy, ¡pero es completamente inútil para detectar a los que sí se van!

````python
# Cálculo de métricas adicionales
acc = accuracy_score(y_test, y_pred) # exactitud general del modelo
print(f"Accuracy: {acc:.2f}, que representa la proporción de predicciones correctas sobre el total de predicciones realizadas.")
````

### 3.3 Precisión (Precision) <a id='3.3-precision'></a>

De todas las veces que el modelo predijo 'Positivo' (Churn), ¿qué porcentaje fue correcto?

$$ \text{Precisión} = \frac{\text{TP}}{\text{TP} + \text{FP}} $$

**Cuándo es importante**: Cuando el coste de un **Falso Positivo** es alto.
- **Ejemplo**: En un filtro de spam, no quieres que un email importante (No Spam) sea clasificado como Spam (Falso Positivo). Quieres que las predicciones de "Spam" sean muy precisas.

**⚠️ Caso especial - Precisión = 0**: Si el modelo **NUNCA predice la clase positiva** (TP + FP = 0), entonces la precisión no puede calcularse o se considera 0. En nuestro ejemplo, el modelo nunca predice "Churn", por lo que:
- No hay verdaderos positivos (TP = 0)
- No hay falsos positivos (FP = 0)
- Por tanto: Precisión = 0/0 = **indefinido → se reporta como 0**

````python
# Precisión del modelo
prec = precision_score(y_test, y_pred, zero_division=0)  # zero_division=0 trata el caso 0/0 como 0
print(f"Precisión: {prec:.2f}, que indica la proporción de predicciones positivas correctas entre todas las predicciones positivas realizadas por el modelo.")
print("Valores de precisión bajos indican que muchas predicciones positivas son incorrectas.")
print(f"\n💡 En este caso: Precisión = 0 porque el modelo NUNCA predice 'Churn' (no hay TP ni FP)")
````

### 3.4 Recall (Sensibilidad o Exhaustividad) <a id='3.4-recall'></a>

De todos los casos que eran realmente 'Positivo' (Churn), ¿qué porcentaje fue capaz de encontrar el modelo?

$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}} $$

**Cuándo es importante**: Cuando el coste de un **Falso Negativo** es alto.
- **Ejemplo**: En el diagnóstico de cáncer, no quieres que un paciente enfermo sea diagnosticado como sano (Falso Negativo). Quieres encontrar (hacer "recall" de) todos los casos positivos posibles.

**⚠️ Caso especial - Recall = 0**: Si el modelo **NUNCA predice la clase positiva**, entonces no detecta ningún caso real positivo:
- No hay verdaderos positivos (TP = 0)
- Hay muchos falsos negativos (FN = casos positivos no detectados)
- Por tanto: Recall = 0 / (0 + FN) = 0/18 = **0.00**

En nuestro ejemplo: El modelo predijo "NO Churn" para todos los 60 clientes. De los 18 que realmente hicieron churn, detectó 0. **Recall = 0%** = modelo completamente inútil para detectar churn.

````python
# Recall del modelo, es decir, sensibilidad
rec = recall_score(y_test, y_pred)
print(f"Recall: {rec:.2f}, que mide la capacidad del modelo para identificar correctamente los casos positivos reales.")
print("Valores de recall bajos indican que muchos casos positivos reales no fueron detectados por el modelo.")
print(f"\n💡 En este caso: Recall = 0 porque el modelo predice 'NO Churn' para TODOS los clientes")
print(f"   De los {y_test.sum()} clientes que SÍ hicieron churn, detectó 0. Esto es un FALLO crítico.")
````

### 3.5 El Dilema Precisión vs. Recall <a id='3.5-dilema'></a>

A menudo, existe un **trade-off** (intercambio) entre precisión y recall. 

- Si quieres **aumentar el recall** (encontrar a todos los que hacen churn), puedes ser menos estricto. Esto hará que clasifiques a más gente como 'Churn', pero también cometerás más Falsos Positivos, **bajando la precisión**.
- Si quieres **aumentar la precisión** (estar muy seguro de que cada predicción de 'Churn' es correcta), serás más estricto. Esto hará que te dejes fuera a algunos casos dudosos, cometiendo más Falsos Negativos y **bajando el recall**.

![Precision-Recall Tradeoff](imagenes/Precision-Recall-Tradeoff)
Generalmente al aumentar el recall, la precisión tiende a disminuir. Esto refleja el equilibrio típico al ajustar umbrales de decisión en modelos de clasificación
 
 * Cuando el umbral de decisión es alto → la precisión suele ser alta pero el recall bajo.

 * Cuando el umbral es bajo → el recall mejora pero la precisión cae.
La elección depende del problema de negocio.

### 3.6 F1-Score: La Media Armónica <a id='3.6-f1'></a>

El F1-Score es una métrica que combina la precisión y el recall en un solo número. Es la **media armónica** de ambos, lo que significa que penaliza más los valores extremos. 

$$ F_1 = 2 \cdot \frac{\text{Precisión} \cdot \text{Recall}}{\text{Precisión} + \text{Recall}} $$

**Cuándo es útil**: Cuando quieres un buen equilibrio entre Precisión y Recall, o cuando tienes un **conjunto de datos desbalanceado**.

````python
# F1-Score del modelo, la media armónica entre precisión y recall
f1 = f1_score(y_test, y_pred, zero_division=0)  # zero_division=0 trata el caso especial cuando precision y recall son 0
print(f"F1-Score: {f1:.2f}, que es la media armónica entre precisión y recall, proporcionando una medida balanceada del rendimiento del modelo.")
print("Valores bajos de F1-Score indican un pobre equilibrio entre precisión y recall.")
print(f"\n💡 En este caso: F1-Score = 0 porque ambas (Precisión y Recall) son 0")
print(f"   La fórmula es: F1 = 2 * (Precisión * Recall) / (Precisión + Recall)")
print(f"   Si Precisión=0 y Recall=0 → F1 = 2 * (0 * 0) / (0 + 0) = 0 (indefinido, se reporta como 0)")
````

### 3.7 Reporte de Clasificación de Scikit-learn <a id='3.7-reporte'></a>

Scikit-learn ofrece una función muy útil que calcula todas estas métricas de una vez.

#### ⚠️ Advertencia de Precisión (UndefinedMetricWarning)

Si ejecutas `classification_report()` sin el parámetro `zero_division`, probablemente verás una advertencia como esta:

```
UndefinedMetricWarning: Precision is ill-defined and being set to 0.0 in labels with no predicted samples. 
Use `zero_division` parameter to control this behavior.
```

**¿Por qué aparece esta advertencia?**

Recuerda que Precisión = TP / (TP + FP). Cuando el modelo **nunca predice la clase positiva**, tenemos:
- TP = 0
- FP = 0
- Por tanto: Precisión = 0 / 0 = **indefinido** (división por cero en matemáticas)

Scikit-learn te avisa de este problema con la advertencia `UndefinedMetricWarning` (advertencia de métrica indefinida).

**¿Cómo se arregla?**

Simplemente añade el parámetro `zero_division=0` a la función:

```python
# ❌ SIN el parámetro (genera advertencia)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'])

# ✅ CON el parámetro (sin advertencia, explícito)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'], zero_division=0)
```

**¿Qué hace `zero_division=0`?**

Le dice a scikit-learn: "Si encuentras una división por cero (como 0/0), simplemente reporta 0 en lugar de advertir".

**Opciones disponibles:**
- `zero_division=0` → Reporta 0 cuando hay división por cero
- `zero_division=1` → Reporta 1 cuando hay división por cero
- `zero_division='warn'` → Muestra la advertencia (comportamiento por defecto)

````python
# Reporte de Clasificación, sirve para ver todas las métricas juntas
# IMPORTANTE: Usamos zero_division=0 para evitar advertencias (UndefinedMetricWarning)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'], zero_division=0)
print("Reporte de Clasificación:")
print(report)
````

## 🚨 ¿POR QUÉ PRECISIÓN, RECALL Y F1-SCORE SON 0?

Esta es una pregunta muy importante y muy común. Veamos qué pasó en nuestro modelo:

### 📊 El Problema: Modelo Que NUNCA Predice la Clase Positiva

En nuestra matriz de confusión:
- **TP (Verdadero Positivo) = 0**: El modelo NUNCA predijo "Churn" correctamente
- **FP (Falso Positivo) = 0**: El modelo NUNCA predijo "Churn" (ni siquiera incorrectamente)
- **FN (Falso Negativo) = 18**: El modelo dijo "NO Churn" pero había 18 que SÍ hicieron churn

**En resumen**: El modelo predijo "NO Churn" para los 60 clientes del conjunto de prueba. 0 predicciones positivas.

### 🔍 Por Eso las Métricas Valen 0:

| Métrica | Fórmula | En Este Caso | Resultado |
|:---|:---|:---|:---|
| **Precisión** | TP / (TP + FP) | 0 / (0 + 0) | **0/0 → 0** |
| **Recall** | TP / (TP + FN) | 0 / (0 + 18) | **0/18 → 0** |
| **F1-Score** | 2 × (Precisión × Recall) / (Precisión + Recall) | 2 × (0 × 0) / (0 + 0) | **0 → 0** |

**Nota**: Cuando aparecen 0/0 (indefinido matemáticamente), scikit-learn lo reporta como 0 por defecto.

### 💡 Significado Pedagógico:

- **Precisión = 0** → El modelo nunca predijo "Churn", así que no podemos saber si sus predicciones positivas son precisas.
- **Recall = 0** → El modelo no detectó NINGUNO de los 18 clientes que realmente hicieron churn. Es un fallo catastrófico en un contexto de negocio real.
- **F1-Score = 0** → El modelo es completamente inútil para detectar churn.

### 🎯 ¿Qué Significa Esto?

El modelo aprendió a predecir "NO Churn" para casi todos los casos. En la práctica, esto significa:

1. **El modelo está sesgado** hacia la clase mayoritaria (No Churn).
2. **Los datos probablemente están desbalanceados** (más clientes leales que churners).
3. **El modelo necesita ser corregido** usando técnicas como:
   - Ajustar los pesos de las clases (`class_weight='balanced'`)
   - Usar técnicas de remuestreo (sobremuestreo de la minoría o submuestreo de la mayoría)
   - Cambiar el umbral de decisión
   - Usar métricas de evaluación diferentes (ROC-AUC, F1-Score en lugar de Accuracy)

**Lección importante**: El Accuracy (70% en este caso) puede ser engañoso. Nuestro modelo tiene un 70% de accuracy, pero un Recall de 0%... ¡no detecta ningún churn!

### 🔧 ¿Cómo Se Arregla Este Problema?

Existen varias técnicas para resolver este problema de desbalanceo de clases. Aquí están las más comunes:

#### **Técnica 1: Ajustar los Pesos de las Clases** (Lo más fácil)

Cuando entrenamos el modelo, podemos decirle: "oye, la clase positiva (Churn) es importante, dale más peso".

```python
model = LogisticRegression(
    class_weight='balanced',  # ← Esta línea es clave
    random_state=42
)
model.fit(X_train, y_train)
```

**🔧 Uso Sencillo:**
1. Solo añade `class_weight='balanced'` en el constructor del modelo
2. Entrena normalmente
3. El modelo automáticamente asignará más importancia a los casos de la clase minoritaria (Churn)
4. ¡Sin necesidad de cambiar los datos!

**Ventaja**: Una línea de código, muy fácil.
**Desventaja**: Puede no ser suficiente en casos muy desbalanceados.

---

#### **Técnica 2: Remuestreo - Sobremuestreo (Oversampling)**

Duplicar los casos de la clase minoritaria (Churn) para que haya más durante el entrenamiento.

```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)
X_train_balanced, y_train_balanced = oversampler.fit_resample(X_train, y_train)

model = LogisticRegression(random_state=42)
model.fit(X_train_balanced, y_train_balanced)
```

**🔧 Uso Sencillo:**
1. Crea un objeto `RandomOverSampler()`
2. Aplica `.fit_resample()` en los datos de entrenamiento (genera copias de las muestras minoritarias)
3. Entrena el modelo con los datos rebalanceados
4. El modelo ve más ejemplos de churn durante el aprendizaje

**Ventaja**: El modelo ve más ejemplos de churn.
**Desventaja**: Podría causar overfitting (memorizar los datos).

---

#### **Técnica 3: Remuestreo - Submuestreo (Undersampling)**

Reducir los casos de la clase mayoritaria (No Churn) para equilibrar las clases.

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=42)
X_train_balanced, y_train_balanced = undersampler.fit_resample(X_train, y_train)

model = LogisticRegression(random_state=42)
model.fit(X_train_balanced, y_train_balanced)
```

**🔧 Uso Sencillo:**
1. Crea un objeto `RandomUnderSampler()`
2. Aplica `.fit_resample()` en los datos de entrenamiento (elimina aleatoriamente muestras mayoritarias)
3. Entrena el modelo con los datos rebalanceados
4. El modelo entrena con menos datos pero más equilibrados

**Ventaja**: Entrena más rápido con menos datos.
**Desventaja**: Perdemos información de la clase mayoritaria.

---

#### **Técnica 4: Cambiar la Métrica de Evaluación**

En lugar de usar **Accuracy**, usar **F1-Score** o **ROC-AUC** que son robustos al desbalanceo.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score, make_scorer

# Buscar hiperparámetros optimizando F1-Score, no Accuracy
scorer = make_scorer(f1_score)
grid_search = GridSearchCV(model, param_grid, scoring=scorer)
```

**🔧 Uso Sencillo:**
1. En lugar de usar `scoring='accuracy'` en GridSearchCV, usa `scoring='f1'` o `scoring='roc_auc'`
2. Al evaluar, reemplaza `accuracy_score()` con `f1_score()` o `roc_auc_score()`
3. El modelo ahora optimiza hacia lo que realmente importa (detectar ambas clases), no solo acertar el máximo

**Ventaja**: Guía al modelo hacia lo que realmente importa.
**Beneficio Pedagógico**: Puedes usarlo con el modelo original, no necesitas cambiar datos ni arquitectura.

---

#### **Técnica 5: Ajustar el Umbral de Decisión**

Por defecto, el modelo predice "Churn" si la probabilidad es > 0.5. Podemos bajar ese umbral:

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]

# En lugar de y_pred = (y_pred_proba > 0.5)
# Usamos un umbral más bajo
y_pred_custom = (y_pred_proba > 0.3)  # Ahora es más "agresivo" prediciendo churn
```

**🔧 Uso Sencillo:**
1. Obtén probabilidades: `y_proba = model.predict_proba(X_test)[:, 1]`
2. Crea predicciones personalizadas: `y_pred = (y_proba > umbral)` donde umbral puede ser 0.3, 0.4, etc.
3. Experimenta con diferentes umbrales (0.3, 0.4, 0.5, 0.6) para encontrar el balance entre precisión y recall
4. Cuanto más bajo el umbral, más agresivo prediciendo churn (mayor recall)

**Ventaja**: Simple y efectivo para ajustar la sensibilidad.
**Desventaja**: Requiere experimentación para encontrar el umbral óptimo.

## 📈 5. Curva ROC y AUC <a id='5-roc'></a>

La **Curva ROC (Receiver Operating Characteristic)** es una de las herramientas de evaluación más importantes para clasificadores binarios.

### 5.1 Visualizando el Rendimiento de un Clasificador <a id='5.1-visualizacion'></a>


La curva ROC visualiza dos métricas:
- **Tasa de Verdaderos Positivos (TPR)** en el eje Y. Es lo mismo que el **Recall**.
- **Tasa de Falsos Positivos (FPR)** en el eje X. Es la proporción de negativos reales que fueron incorrectamente clasificados como positivos ( `FP / (FP + TN)` ).

Un buen modelo es aquel que maximiza la TPR mientras minimiza la FPR. Esto se traduce en una curva que se acerca lo más posible a la **esquina superior izquierda** del gráfico.

**AUC (Area Under the Curve)**: Es el área bajo la curva ROC. 
- **AUC = 1**: Clasificador perfecto.
- **AUC = 0.5**: Clasificador inútil (azar, como lanzar una moneda).
- **AUC < 0.5**: Clasificador peor que el azar.

El AUC es muy útil porque resume el rendimiento del modelo en un solo número, independientemente del umbral de clasificación elegido.

````python
from sklearn.metrics import roc_curve, roc_auc_score

# Necesitamos las probabilidades de la clase positiva, no solo las predicciones
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcular el AUC
auc = roc_auc_score(y_test, y_pred_proba)

print(f"AUC: {auc:.2f}")

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Azar')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
print("Curva ROC graficada.")
print("Este gráfico muestra la capacidad del modelo para distinguir entre las clases positivas y negativas a diferentes umbrales de decisión.")
````

## 📌 4. Adelanto: ¿Qué es la Regresión Logística? <a id='4-adelanto-regresion-logistica'></a>

Probablemente te has dado cuenta de algo: en todo este módulo hemos estado usando un modelo llamado **LogisticRegression** (Regresión Logística) para realizar predicciones de clasificación. Aunque profundizaremos mucho en este modelo en el próximo módulo, aquí te damos un adelanto conceptual.

### 🤔 ¿Por qué se llama "Regresión" si estamos clasificando?

Es una buena pregunta. El nombre es histórico y puede ser confuso. Aunque su nombre contenga la palabra "regresión", la **Regresión Logística es en realidad un modelo de CLASIFICACIÓN**, no de regresión.

### 🧮 ¿Cómo funciona conceptualmente?

**Regresión Logística** es, en esencia:

1. **Regresión Lineal + Función Especial**: Toma una regresión lineal (como la que vimos en el tema anterior) y aplica una función especial llamada **función logística** o **sigmoid**.

2. **De números continuos a probabilidades**: 
   - Una regresión lineal te da un número cualquiera (puede ser -100, +1000, etc.)
   - La función logística transforma ese número en una **probabilidad** (un número entre 0 y 1)
   
3. **De probabilidades a clases**:
   - Si la probabilidad es > 0.5 → Predice clase **1** (Sí, Churn)
   - Si la probabilidad es ≤ 0.5 → Predice clase **0** (No, No Churn)

### 📊 Ejemplo Visual Conceptual

```
Entrada: Antigüedad=6 meses, Gasto=150€
         ↓
    [Regresión Lineal]
         ↓
    Valor numérico: 3.2
         ↓
    [Función Logística/Sigmoid]
         ↓
    Probabilidad: 0.96 (96% de probabilidad de churn)
         ↓
    Predicción Final: SÍ (1) - porque 0.96 > 0.5
```

| **Clasificación** | **Datos desbalanceados** (ej: fraude, enfermedad rara) | **AUC**, **F1-Score**, **Precision/Recall** |
| | El coste de los **Falsos Negativos** es alto (ej: diagnóstico médico) | **Recall** (Sensibilidad) |
| | El coste de los **Falsos Positivos** es alto (ej: filtro de spam) | **Precisión** |
| | Datos balanceados y coste de errores simétrico | **Accuracy** (pero siempre mira las otras) |
| **Regresión** | Quieres una métrica **fácil de explicar** a negocio | **MAE** |
| | Los **errores grandes son muy malos** y deben ser penalizados | **RMSE** (o MSE) |
| | Quieres saber **qué tan bien se ajusta tu modelo** a los datos | **R²** |

**Consejo Pro**: Nunca te fíes de una sola métrica. Analiza varias para tener una visión completa del rendimiento de tu modelo.
