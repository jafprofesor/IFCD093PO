## üîÑ Comparaci√≥n: Regresi√≥n vs Clasificaci√≥n

En el tema anterior, trabajamos con **Regresi√≥n Lineal**. Ahora vamos a cambiar de enfoque con **Clasificaci√≥n** (regresi√≥n log√≠stica). Es importante entender la diferencia fundamental:

### üìä ¬øQu√© es Regresi√≥n?

En **regresi√≥n**, el modelo **predice un valor num√©rico continuo**. 

**Ejemplo del tema anterior: Predicci√≥n de Salarios**
- **Entrada**: A√±os de experiencia (n√∫mero continuo: 5.0, 6.5, 15.3...)
- **Salida**: Salario predicho (n√∫mero continuo: 45,250‚Ç¨, 72,300‚Ç¨...)
- **Pregunta que hace el modelo**: "¬øCu√°l es el salario probable para un empleado con X a√±os de experiencia?"
- **Tipo de respuesta**: Un n√∫mero espec√≠fico

La salida es una **cantidad**, algo que se mide en una escala continua (euros, metros, kilos, etc.).

### üéØ ¬øQu√© es Clasificaci√≥n?

En **clasificaci√≥n**, el modelo **predice la categor√≠a o clase a la que pertenece un objeto**.

**Ejemplo que veremos ahora: Predicci√≥n de Churn (Abandono)**
- **Entrada**: Antig√ºedad del cliente, gasto mensual (n√∫meros continuos)
- **Salida**: ¬øAbandona el cliente? ‚Üí **S√ç (1)** o **NO (0)**
- **Pregunta que hace el modelo**: "¬øEste cliente va a abandonar el servicio o no?"
- **Tipo de respuesta**: Una categor√≠a, no un n√∫mero exacto

### üìà Comparaci√≥n Visual

| Caracter√≠stica | **Regresi√≥n** | **Clasificaci√≥n** |
|:---|:---|:---|
| **¬øQu√© predice?** | Valores continuos (n√∫meros reales) | Categor√≠as discretas (clases) |
| **Ejemplos de salida** | 45,250‚Ç¨, 72,300‚Ç¨, 55,000‚Ç¨... | Churn=S√ç, Churn=NO |
| **Gr√°ficamente** | Una l√≠nea o curva suave | Puntos agrupados por clase |
| **Pregunta t√≠pica** | "¬øCu√°nto?" | "¬øCu√°l?" o "¬øEs...?" |
| **Ejemplo real** | Precio de casa, temperatura, ventas | Email spam/no-spam, diagn√≥stico m√©dico, cliente leal/infiel |
| **Funci√≥n del modelo** | ùë¶ = ùúÉ‚ÇÄ + ùúÉ‚ÇÅùë• (Regresi√≥n Lineal) | ùëù = 1/(1+e‚ÅªÀ¢) (Regresi√≥n Log√≠stica) |

### ü§î ¬øC√≥mo Evaluamos Cada Uno?

Por esta diferencia fundamental, **no podemos usar las mismas m√©tricas** para regresi√≥n y clasificaci√≥n:

- **En Regresi√≥n**: Medimos qu√© tan cerca est√° la predicci√≥n del valor real. Usamos **MAE, MSE, RMSE, R¬≤**.
  - Pregunta: "¬øCu√°nto se equivoc√≥ el modelo en euros?"

- **En Clasificaci√≥n**: No tiene sentido preguntar "cu√°nto se equivoc√≥" (¬øen qu√© unidades?). En cambio, contamos:
  - ¬øCu√°ntas predicciones fueron correctas? (**Accuracy**)
  - ¬øCu√°ntos casos positivos detectamos correctamente? (**Recall**)
  - ¬øQu√© tan seguros estamos de nuestras predicciones positivas? (**Precisi√≥n**)
  - Usamos la **Matriz de Confusi√≥n, Precision, Recall, F1-Score, AUC-ROC**.

### üí° Ejemplo Pr√°ctico del Cambio

**Modelo de Regresi√≥n:**
```

Output: 87,500‚Ç¨ ‚Üê Predicci√≥n exacta de un n√∫mero

```

**Modelo de Clasificaci√≥n:**
```

Output: Churn = S√ç (1) ‚Üê Predicci√≥n de una clase
Real: Churn = S√ç (1)
Resultado: ‚úÖ Acierto (Verdadero Positivo)

Input: Cliente con gasto mensual de 200‚Ç¨, antig√ºedad de 24 meses
Output: Churn = NO (0) ‚Üê Predicci√≥n de una clase
Real: Churn = S√ç (1)
Resultado: ‚ùå Error (Falso Negativo)
```

## ü§î 1. La Importancia de la M√©trica Correcta <a id='1-importancia'></a>

Imagina dos escenarios:

1.  **Diagn√≥stico de C√°ncer**: Un modelo que predice si un tumor es maligno.
2.  **Recomendaci√≥n de V√≠deos**: Un modelo que sugiere v√≠deos en YouTube.

- En la **recomendaci√≥n de v√≠deos**, un **Falso Positivo** (recomendar un v√≠deo que no te gusta) es molesto, pero no grave. Queremos maximizar los aciertos, pero un error no tiene consecuencias serias.

La elecci√≥n de la m√©trica de evaluaci√≥n depende directamente del **objetivo de negocio** y de las **consecuencias de los diferentes tipos de errores**. No existe "la mejor m√©trica" para todos los casos.

## üìä 2. M√©tricas para Problemas de Regresi√≥n <a id='2-regresion'></a>

En regresi√≥n, no predecimos categor√≠as, sino valores continuos. Por lo tanto, las m√©tricas se basan en medir la **distancia** o el **error** entre el valor real (`y`) y el valor predicho (`≈∑`).

Para explicar estas m√©tricas, usaremos un problema de regresi√≥n: predecir el precio de casas bas√°ndonos en su tama√±o.

````python
# Preparaci√≥n: Entrenemos un modelo de regresi√≥n para tener predicciones que evaluar
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Datos sint√©ticos de precios de casas
np.random.seed(42)
tama√±o = np.random.randint(50, 200, 100)
precio_real = tama√±o * 1500 + np.random.normal(0, 25000, 100)
df_casas = pd.DataFrame({'Tama√±o': tama√±o, 'Precio': precio_real})

X_casas = df_casas[['Tama√±o']]
y_casas = df_casas['Precio']

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_casas, y_casas, test_size=0.2, random_state=42)

reg_model = LinearRegression()
reg_model.fit(X_train_c, y_train_c)
y_pred_c = reg_model.predict(X_test_c)

print("Modelo de regresi√≥n entrenado y predicciones listas para evaluar.")
````

### 2.1 Error Absoluto Medio (MAE) <a id='2.1-mae'></a>

Es el promedio de las diferencias absolutas entre los valores reales y los predichos.

$$ \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$

- **Interpretaci√≥n**: "En promedio, nuestro modelo se equivoca en X unidades".
- **Ventaja**: F√°cil de interpretar porque est√° en las mismas unidades que la variable objetivo.
- **Desventaja**: No penaliza m√°s los errores grandes.

````python
# C√°lculo de MAE (Mean Absolute Error - Error Absoluto Medio)
mae = mean_absolute_error(y_test_c, y_pred_c)
print(f"MAE: {mae:.2f}‚Ç¨")
````

### 2.2 Error Cuadr√°tico Medio (MSE) <a id='2.2-mse'></a>

Es el promedio de los errores al cuadrado.

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

- **Ventaja**: Penaliza mucho m√°s los errores grandes que los peque√±os (debido al cuadrado). Es muy √∫til si los errores grandes son especialmente indeseables.
- **Desventaja**: Las unidades est√°n al cuadrado (ej: ‚Ç¨¬≤), lo que lo hace dif√≠cil de interpretar.

````python
# C√°lculo de MSE (Mean Squared Error - Error Cuadr√°tico Medio)
mse = mean_squared_error(y_test_c, y_pred_c)
print(f"MSE: {mse:.2f}‚Ç¨¬≤")
````

### 2.3 Ra√≠z del Error Cuadr√°tico Medio (RMSE) <a id='2.3-rmse'></a>

Es simplemente la ra√≠z cuadrada del MSE. 

$$ \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} $$

- **Ventaja**: Combina lo mejor del MAE y el MSE. Penaliza los errores grandes (como el MSE) pero est√° en las mismas unidades que la variable objetivo (como el MAE), haci√©ndolo interpretable.
- **Es una de las m√©tricas m√°s utilizadas en regresi√≥n.**

````python
# C√°lculo del RMSE (Raiz del error cuadr√°tico medio)
rmse = np.sqrt(mse)
print(f"RMSE: {rmse:.2f}‚Ç¨")
````

### 2.4 Coeficiente de Determinaci√≥n (R¬≤) <a id='2.4-r2'></a>

R¬≤ (R-cuadrado) mide la **proporci√≥n de la varianza** en la variable dependiente que es predecible a partir de las variables independientes.

- **Interpretaci√≥n**: "Nuestro modelo explica el X% de la variabilidad en los precios de las casas".
- **Rango**: Va de -‚àû a 1.
  - **R¬≤ = 1**: El modelo explica toda la variabilidad (predicciones perfectas).
  - **R¬≤ = 0**: El modelo no explica nada de la variabilidad (es tan bueno como predecir siempre la media).
  - **R¬≤ < 0**: El modelo es peor que predecir la media.

**Cuidado**: R¬≤ tiende a aumentar si a√±ades m√°s variables al modelo, incluso si no son √∫tiles. Para eso existe el **R¬≤ ajustado**, que penaliza la adici√≥n de variables innecesarias.

````python
# C√°lculo del R2 (Coeficiente de Determinaci√≥n)
r2 = r2_score(y_test_c, y_pred_c)
print(f"R¬≤: {r2:.2f}")
print(f"Esto significa que nuestro modelo explica aproximadamente el {r2*100:.0f}% de la variabilidad en los precios de las casas.")
````

## üéØ 3. M√©tricas para Problemas de Clasificaci√≥n <a id='3-clasificacion'></a>

Para explicar estas m√©tricas, usaremos un problema de clasificaci√≥n binaria: predecir si un cliente abandonar√° o no una compa√±√≠a (Churn).

- **Clase Positiva (1)**: El cliente S√ç abandona (Churn).
- **Clase Negativa (0)**: El cliente NO abandona (No Churn).

````python
# Preparaci√≥n: Entrenemos un modelo simple para tener predicciones que evaluar
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split # Divisi√≥n de datos
from sklearn.linear_model import LogisticRegression # Modelo de regresi√≥n log√≠stica
# M√©tricas de clasificaci√≥n
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report 

# Datos sint√©ticos de Churn
# Aqu√≠ generamos un conjunto de datos sint√©ticos para simular el problema de churn
np.random.seed(42) # Semilla para reproducibilidad
data = {
    'antiguedad': np.random.randint(1, 60, 200),
    'gasto_mensual': np.random.uniform(20, 200, 200),
    'churn': np.random.randint(0, 2, 200)
}
# Forzar un desbalanceo para que las m√©tricas sean m√°s interesantes
data['churn'][:140] = 0 # 70% no churn
data['churn'][140:] = 1 # 30% churn

# Crear DataFrame con los datos
df_churn = pd.DataFrame(data).sample(frac=1, random_state=42).reset_index(drop=True)

# Dividir en conjunto de entrenamiento y prueba
X = df_churn[['antiguedad', 'gasto_mensual']] # Caracter√≠sticas 
y = df_churn['churn'] # Etiqueta objetivo

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Entrenar un modelo simple de regresi√≥n log√≠stica
model = LogisticRegression(random_state=42) # Modelo de regresi√≥n log√≠stica
model.fit(X_train, y_train) # Entrenamiento del modelo
y_pred = model.predict(X_test) # Predicciones del modelo

print("Modelo entrenado y predicciones listas para evaluar.")
````

### 3.1 La Matriz de Confusi√≥n: La Base de Todo <a id='3.1-matriz'></a>

La matriz de confusi√≥n es una tabla que resume el rendimiento de un modelo de clasificaci√≥n. Compara las etiquetas reales con las predicciones del modelo.

|                | **Predicci√≥n: NO (0)** | **Predicci√≥n: S√ç (1)** |
|----------------|------------------------|------------------------|
| **Real: NO (0)** | Verdadero Negativo (TN) | Falso Positivo (FP)    |
| **Real: S√ç (1)** | Falso Negativo (FN)    | Verdadero Positivo (TP) |

- **Verdadero Positivo (TP)**: El modelo predijo 'Churn' y el cliente realmente hizo 'Churn'. ¬°Acierto!
- **Verdadero Negativo (TN)**: El modelo predijo 'No Churn' y el cliente realmente 'No Churn'. ¬°Acierto!
- **Falso Positivo (FP) / Error Tipo I**: El modelo predijo 'Churn', pero el cliente se qued√≥. (Ej: Ofrecer un descuento innecesariamente).
- **Falso Negativo (FN) / Error Tipo II**: El modelo predijo 'No Churn', pero el cliente se fue. (Ej: Perder un cliente por no actuar a tiempo). **¬°Este suele ser el m√°s costoso!**

````python
import seaborn as sns # para visualizaci√≥n
import matplotlib.pyplot as plt # para visualizaci√≥n

cm = confusion_matrix(y_test, y_pred) # calcular la matriz de confusi√≥n
# Le pasamos y_test (etiquetas reales) y y_pred (etiquetas predichas por el modelo)
print("Matriz de Confusi√≥n:")
print(cm)

# Visualizaci√≥n de la Matriz de Confusi√≥n
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', # A√±adido etiquetas para mayor claridad
            xticklabels=['Pred. No Churn', 'Pred. Churn'], 
            yticklabels=['Real No Churn', 'Real Churn'])
plt.title('Matriz de Confusi√≥n')
plt.ylabel('Etiqueta Real')
plt.xlabel('Etiqueta Predicha')
plt.show()
````

#### üìä Interpretaci√≥n de Tu Matriz de Confusi√≥n

En tu gr√°fico aparecen estos n√∫meros exactos:

```
                 Pred. No Churn  |  Pred. Churn
Real No Churn        42          |      0
Real Churn           18          |      0
```

**Desglose por cuadrante:**

| Posici√≥n | Valor | Nombre | Significado |
|:---|:---|:---|:---|
| **Superior Izquierda** | **42** | TN (Verdadero Negativo) | ‚úÖ 42 clientes que NO abandonaron ‚Üí Modelo predijo bien |
| **Superior Derecha** | **0** | FP (Falso Positivo) | ‚úÖ 0 clientes incorrectamente predichos como "Churn" |
| **Inferior Izquierda** | **18** | FN (Falso Negativo) | ‚ùå‚ùå 18 clientes que S√ç abandonaron ‚Üí Pero el modelo dijo "NO Churn" |
| **Inferior Derecha** | **0** | TP (Verdadero Positivo) | ‚ùå 0 clientes en churn detectados correctamente |

#### üö® Diagn√≥stico Cr√≠tico:

**El modelo es in√∫til para detectar churn:**
- ‚úÖ **Lo hace bien**: Identifica clientes que NO se van (42 casos)
- ‚ùå **Lo hace mal**: Nunca detecta clientes que S√ç se van
- ‚ùå **Problema**: 18 clientes reales abandonaron sin que el modelo levantara ninguna alarma

**M√©tricas resultantes:**
- **Recall = 0%** ‚Üí No detecta NING√öN caso de churn real
- **Precisi√≥n = indefinida** ‚Üí No puede calcularse (nunca predice positivos)
- **El modelo predice "NO Churn" para TODOS los clientes**

````python
# An√°lisis detallado de la matriz
tn, fp, fn, tp = cm.ravel()

print(f"üìä AN√ÅLISIS DETALLADO DE LA MATRIZ:")
print(f"‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")
print(f"TN (Superior Izq):        {tn} - Clientes leales bien identificados ‚úÖ")
print(f"FP (Superior Der):        {fp} - Falsos positivos (ninguno) ‚úÖ")
print(f"FN (Inferior Izq):        {fn} - CLIENTES PERDIDOS sin detectar ‚ùå‚ùå")
print(f"TP (Inferior Der):        {tp} - Churns detectados (ninguno) ‚ùå")
print(f"‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")
print(f"\nüìä TOTALES:")
print(f"Clientes que NO se fueron: {tn + fp}")
print(f"Clientes que S√ç se fueron: {fn + tp}")
print(f"Total de predicciones: {tn + fp + fn + tp}")

if (fn + tp) > 0:
    deteccion = 100*tp/(fn+tp)
else:
    deteccion = 0
    
print(f"\nüö® PROBLEMA CR√çTICO:")
print(f"De los {fn + tp} clientes que abandonaron: {tp} fueron detectados ({deteccion:.1f}%)")
print(f"El modelo FALL√ì en {fn} casos importantes (clientes perdidos)")
````

### 3.2 Accuracy (Exactitud) <a id='3.2-accuracy'></a>

Es la m√©trica m√°s intuitiva: ¬øQu√© porcentaje de predicciones fueron correctas?

$$ \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}} = \frac{\text{Predicciones Correctas}}{\text{Total de Predicciones}} $$

**Cuidado**: El accuracy puede ser enga√±oso en **conjuntos de datos desbalanceados**. Si el 99% de tus clientes no hacen churn, un modelo que *siempre* predice "No Churn" tendr√° un 99% de accuracy, ¬°pero es completamente in√∫til para detectar a los que s√≠ se van!

````python
# C√°lculo de m√©tricas adicionales
acc = accuracy_score(y_test, y_pred) # exactitud general del modelo
print(f"Accuracy: {acc:.2f}, que representa la proporci√≥n de predicciones correctas sobre el total de predicciones realizadas.")
````

### 3.3 Precisi√≥n (Precision) <a id='3.3-precision'></a>

De todas las veces que el modelo predijo 'Positivo' (Churn), ¬øqu√© porcentaje fue correcto?

$$ \text{Precisi√≥n} = \frac{\text{TP}}{\text{TP} + \text{FP}} $$

**Cu√°ndo es importante**: Cuando el coste de un **Falso Positivo** es alto.
- **Ejemplo**: En un filtro de spam, no quieres que un email importante (No Spam) sea clasificado como Spam (Falso Positivo). Quieres que las predicciones de "Spam" sean muy precisas.

**‚ö†Ô∏è Caso especial - Precisi√≥n = 0**: Si el modelo **NUNCA predice la clase positiva** (TP + FP = 0), entonces la precisi√≥n no puede calcularse o se considera 0. En nuestro ejemplo, el modelo nunca predice "Churn", por lo que:
- No hay verdaderos positivos (TP = 0)
- No hay falsos positivos (FP = 0)
- Por tanto: Precisi√≥n = 0/0 = **indefinido ‚Üí se reporta como 0**

````python
# Precisi√≥n del modelo
prec = precision_score(y_test, y_pred, zero_division=0)  # zero_division=0 trata el caso 0/0 como 0
print(f"Precisi√≥n: {prec:.2f}, que indica la proporci√≥n de predicciones positivas correctas entre todas las predicciones positivas realizadas por el modelo.")
print("Valores de precisi√≥n bajos indican que muchas predicciones positivas son incorrectas.")
print(f"\nüí° En este caso: Precisi√≥n = 0 porque el modelo NUNCA predice 'Churn' (no hay TP ni FP)")
````

### 3.4 Recall (Sensibilidad o Exhaustividad) <a id='3.4-recall'></a>

De todos los casos que eran realmente 'Positivo' (Churn), ¬øqu√© porcentaje fue capaz de encontrar el modelo?

$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}} $$

**Cu√°ndo es importante**: Cuando el coste de un **Falso Negativo** es alto.
- **Ejemplo**: En el diagn√≥stico de c√°ncer, no quieres que un paciente enfermo sea diagnosticado como sano (Falso Negativo). Quieres encontrar (hacer "recall" de) todos los casos positivos posibles.

**‚ö†Ô∏è Caso especial - Recall = 0**: Si el modelo **NUNCA predice la clase positiva**, entonces no detecta ning√∫n caso real positivo:
- No hay verdaderos positivos (TP = 0)
- Hay muchos falsos negativos (FN = casos positivos no detectados)
- Por tanto: Recall = 0 / (0 + FN) = 0/18 = **0.00**

En nuestro ejemplo: El modelo predijo "NO Churn" para todos los 60 clientes. De los 18 que realmente hicieron churn, detect√≥ 0. **Recall = 0%** = modelo completamente in√∫til para detectar churn.

````python
# Recall del modelo, es decir, sensibilidad
rec = recall_score(y_test, y_pred)
print(f"Recall: {rec:.2f}, que mide la capacidad del modelo para identificar correctamente los casos positivos reales.")
print("Valores de recall bajos indican que muchos casos positivos reales no fueron detectados por el modelo.")
print(f"\nüí° En este caso: Recall = 0 porque el modelo predice 'NO Churn' para TODOS los clientes")
print(f"   De los {y_test.sum()} clientes que S√ç hicieron churn, detect√≥ 0. Esto es un FALLO cr√≠tico.")
````

### 3.5 El Dilema Precisi√≥n vs. Recall <a id='3.5-dilema'></a>

A menudo, existe un **trade-off** (intercambio) entre precisi√≥n y recall. 

- Si quieres **aumentar el recall** (encontrar a todos los que hacen churn), puedes ser menos estricto. Esto har√° que clasifiques a m√°s gente como 'Churn', pero tambi√©n cometer√°s m√°s Falsos Positivos, **bajando la precisi√≥n**.
- Si quieres **aumentar la precisi√≥n** (estar muy seguro de que cada predicci√≥n de 'Churn' es correcta), ser√°s m√°s estricto. Esto har√° que te dejes fuera a algunos casos dudosos, cometiendo m√°s Falsos Negativos y **bajando el recall**.

![Precision-Recall Tradeoff](imagenes/Precision-Recall-Tradeoff)
Generalmente al aumentar el recall, la precisi√≥n tiende a disminuir. Esto refleja el equilibrio t√≠pico al ajustar umbrales de decisi√≥n en modelos de clasificaci√≥n
 
 * Cuando el umbral de decisi√≥n es alto ‚Üí la precisi√≥n suele ser alta pero el recall bajo.

 * Cuando el umbral es bajo ‚Üí el recall mejora pero la precisi√≥n cae.
La elecci√≥n depende del problema de negocio.

### 3.6 F1-Score: La Media Arm√≥nica <a id='3.6-f1'></a>

El F1-Score es una m√©trica que combina la precisi√≥n y el recall en un solo n√∫mero. Es la **media arm√≥nica** de ambos, lo que significa que penaliza m√°s los valores extremos. 

$$ F_1 = 2 \cdot \frac{\text{Precisi√≥n} \cdot \text{Recall}}{\text{Precisi√≥n} + \text{Recall}} $$

**Cu√°ndo es √∫til**: Cuando quieres un buen equilibrio entre Precisi√≥n y Recall, o cuando tienes un **conjunto de datos desbalanceado**.

````python
# F1-Score del modelo, la media arm√≥nica entre precisi√≥n y recall
f1 = f1_score(y_test, y_pred, zero_division=0)  # zero_division=0 trata el caso especial cuando precision y recall son 0
print(f"F1-Score: {f1:.2f}, que es la media arm√≥nica entre precisi√≥n y recall, proporcionando una medida balanceada del rendimiento del modelo.")
print("Valores bajos de F1-Score indican un pobre equilibrio entre precisi√≥n y recall.")
print(f"\nüí° En este caso: F1-Score = 0 porque ambas (Precisi√≥n y Recall) son 0")
print(f"   La f√≥rmula es: F1 = 2 * (Precisi√≥n * Recall) / (Precisi√≥n + Recall)")
print(f"   Si Precisi√≥n=0 y Recall=0 ‚Üí F1 = 2 * (0 * 0) / (0 + 0) = 0 (indefinido, se reporta como 0)")
````

### 3.7 Reporte de Clasificaci√≥n de Scikit-learn <a id='3.7-reporte'></a>

Scikit-learn ofrece una funci√≥n muy √∫til que calcula todas estas m√©tricas de una vez.

#### ‚ö†Ô∏è Advertencia de Precisi√≥n (UndefinedMetricWarning)

Si ejecutas `classification_report()` sin el par√°metro `zero_division`, probablemente ver√°s una advertencia como esta:

```
UndefinedMetricWarning: Precision is ill-defined and being set to 0.0 in labels with no predicted samples. 
Use `zero_division` parameter to control this behavior.
```

**¬øPor qu√© aparece esta advertencia?**

Recuerda que Precisi√≥n = TP / (TP + FP). Cuando el modelo **nunca predice la clase positiva**, tenemos:
- TP = 0
- FP = 0
- Por tanto: Precisi√≥n = 0 / 0 = **indefinido** (divisi√≥n por cero en matem√°ticas)

Scikit-learn te avisa de este problema con la advertencia `UndefinedMetricWarning` (advertencia de m√©trica indefinida).

**¬øC√≥mo se arregla?**

Simplemente a√±ade el par√°metro `zero_division=0` a la funci√≥n:

```python
# ‚ùå SIN el par√°metro (genera advertencia)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'])

# ‚úÖ CON el par√°metro (sin advertencia, expl√≠cito)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'], zero_division=0)
```

**¬øQu√© hace `zero_division=0`?**

Le dice a scikit-learn: "Si encuentras una divisi√≥n por cero (como 0/0), simplemente reporta 0 en lugar de advertir".

**Opciones disponibles:**
- `zero_division=0` ‚Üí Reporta 0 cuando hay divisi√≥n por cero
- `zero_division=1` ‚Üí Reporta 1 cuando hay divisi√≥n por cero
- `zero_division='warn'` ‚Üí Muestra la advertencia (comportamiento por defecto)

````python
# Reporte de Clasificaci√≥n, sirve para ver todas las m√©tricas juntas
# IMPORTANTE: Usamos zero_division=0 para evitar advertencias (UndefinedMetricWarning)
report = classification_report(y_test, y_pred, target_names=['No Churn', 'Churn'], zero_division=0)
print("Reporte de Clasificaci√≥n:")
print(report)
````

## üö® ¬øPOR QU√â PRECISI√ìN, RECALL Y F1-SCORE SON 0?

Esta es una pregunta muy importante y muy com√∫n. Veamos qu√© pas√≥ en nuestro modelo:

### üìä El Problema: Modelo Que NUNCA Predice la Clase Positiva

En nuestra matriz de confusi√≥n:
- **TP (Verdadero Positivo) = 0**: El modelo NUNCA predijo "Churn" correctamente
- **FP (Falso Positivo) = 0**: El modelo NUNCA predijo "Churn" (ni siquiera incorrectamente)
- **FN (Falso Negativo) = 18**: El modelo dijo "NO Churn" pero hab√≠a 18 que S√ç hicieron churn

**En resumen**: El modelo predijo "NO Churn" para los 60 clientes del conjunto de prueba. 0 predicciones positivas.

### üîç Por Eso las M√©tricas Valen 0:

| M√©trica | F√≥rmula | En Este Caso | Resultado |
|:---|:---|:---|:---|
| **Precisi√≥n** | TP / (TP + FP) | 0 / (0 + 0) | **0/0 ‚Üí 0** |
| **Recall** | TP / (TP + FN) | 0 / (0 + 18) | **0/18 ‚Üí 0** |
| **F1-Score** | 2 √ó (Precisi√≥n √ó Recall) / (Precisi√≥n + Recall) | 2 √ó (0 √ó 0) / (0 + 0) | **0 ‚Üí 0** |

**Nota**: Cuando aparecen 0/0 (indefinido matem√°ticamente), scikit-learn lo reporta como 0 por defecto.

### üí° Significado Pedag√≥gico:

- **Precisi√≥n = 0** ‚Üí El modelo nunca predijo "Churn", as√≠ que no podemos saber si sus predicciones positivas son precisas.
- **Recall = 0** ‚Üí El modelo no detect√≥ NINGUNO de los 18 clientes que realmente hicieron churn. Es un fallo catastr√≥fico en un contexto de negocio real.
- **F1-Score = 0** ‚Üí El modelo es completamente in√∫til para detectar churn.

### üéØ ¬øQu√© Significa Esto?

El modelo aprendi√≥ a predecir "NO Churn" para casi todos los casos. En la pr√°ctica, esto significa:

1. **El modelo est√° sesgado** hacia la clase mayoritaria (No Churn).
2. **Los datos probablemente est√°n desbalanceados** (m√°s clientes leales que churners).
3. **El modelo necesita ser corregido** usando t√©cnicas como:
   - Ajustar los pesos de las clases (`class_weight='balanced'`)
   - Usar t√©cnicas de remuestreo (sobremuestreo de la minor√≠a o submuestreo de la mayor√≠a)
   - Cambiar el umbral de decisi√≥n
   - Usar m√©tricas de evaluaci√≥n diferentes (ROC-AUC, F1-Score en lugar de Accuracy)

**Lecci√≥n importante**: El Accuracy (70% en este caso) puede ser enga√±oso. Nuestro modelo tiene un 70% de accuracy, pero un Recall de 0%... ¬°no detecta ning√∫n churn!

### üîß ¬øC√≥mo Se Arregla Este Problema?

Existen varias t√©cnicas para resolver este problema de desbalanceo de clases. Aqu√≠ est√°n las m√°s comunes:

#### **T√©cnica 1: Ajustar los Pesos de las Clases** (Lo m√°s f√°cil)

Cuando entrenamos el modelo, podemos decirle: "oye, la clase positiva (Churn) es importante, dale m√°s peso".

```python
model = LogisticRegression(
    class_weight='balanced',  # ‚Üê Esta l√≠nea es clave
    random_state=42
)
model.fit(X_train, y_train)
```

**üîß Uso Sencillo:**
1. Solo a√±ade `class_weight='balanced'` en el constructor del modelo
2. Entrena normalmente
3. El modelo autom√°ticamente asignar√° m√°s importancia a los casos de la clase minoritaria (Churn)
4. ¬°Sin necesidad de cambiar los datos!

**Ventaja**: Una l√≠nea de c√≥digo, muy f√°cil.
**Desventaja**: Puede no ser suficiente en casos muy desbalanceados.

---

#### **T√©cnica 2: Remuestreo - Sobremuestreo (Oversampling)**

Duplicar los casos de la clase minoritaria (Churn) para que haya m√°s durante el entrenamiento.

```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)
X_train_balanced, y_train_balanced = oversampler.fit_resample(X_train, y_train)

model = LogisticRegression(random_state=42)
model.fit(X_train_balanced, y_train_balanced)
```

**üîß Uso Sencillo:**
1. Crea un objeto `RandomOverSampler()`
2. Aplica `.fit_resample()` en los datos de entrenamiento (genera copias de las muestras minoritarias)
3. Entrena el modelo con los datos rebalanceados
4. El modelo ve m√°s ejemplos de churn durante el aprendizaje

**Ventaja**: El modelo ve m√°s ejemplos de churn.
**Desventaja**: Podr√≠a causar overfitting (memorizar los datos).

---

#### **T√©cnica 3: Remuestreo - Submuestreo (Undersampling)**

Reducir los casos de la clase mayoritaria (No Churn) para equilibrar las clases.

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=42)
X_train_balanced, y_train_balanced = undersampler.fit_resample(X_train, y_train)

model = LogisticRegression(random_state=42)
model.fit(X_train_balanced, y_train_balanced)
```

**üîß Uso Sencillo:**
1. Crea un objeto `RandomUnderSampler()`
2. Aplica `.fit_resample()` en los datos de entrenamiento (elimina aleatoriamente muestras mayoritarias)
3. Entrena el modelo con los datos rebalanceados
4. El modelo entrena con menos datos pero m√°s equilibrados

**Ventaja**: Entrena m√°s r√°pido con menos datos.
**Desventaja**: Perdemos informaci√≥n de la clase mayoritaria.

---

#### **T√©cnica 4: Cambiar la M√©trica de Evaluaci√≥n**

En lugar de usar **Accuracy**, usar **F1-Score** o **ROC-AUC** que son robustos al desbalanceo.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score, make_scorer

# Buscar hiperpar√°metros optimizando F1-Score, no Accuracy
scorer = make_scorer(f1_score)
grid_search = GridSearchCV(model, param_grid, scoring=scorer)
```

**üîß Uso Sencillo:**
1. En lugar de usar `scoring='accuracy'` en GridSearchCV, usa `scoring='f1'` o `scoring='roc_auc'`
2. Al evaluar, reemplaza `accuracy_score()` con `f1_score()` o `roc_auc_score()`
3. El modelo ahora optimiza hacia lo que realmente importa (detectar ambas clases), no solo acertar el m√°ximo

**Ventaja**: Gu√≠a al modelo hacia lo que realmente importa.
**Beneficio Pedag√≥gico**: Puedes usarlo con el modelo original, no necesitas cambiar datos ni arquitectura.

---

#### **T√©cnica 5: Ajustar el Umbral de Decisi√≥n**

Por defecto, el modelo predice "Churn" si la probabilidad es > 0.5. Podemos bajar ese umbral:

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]

# En lugar de y_pred = (y_pred_proba > 0.5)
# Usamos un umbral m√°s bajo
y_pred_custom = (y_pred_proba > 0.3)  # Ahora es m√°s "agresivo" prediciendo churn
```

**üîß Uso Sencillo:**
1. Obt√©n probabilidades: `y_proba = model.predict_proba(X_test)[:, 1]`
2. Crea predicciones personalizadas: `y_pred = (y_proba > umbral)` donde umbral puede ser 0.3, 0.4, etc.
3. Experimenta con diferentes umbrales (0.3, 0.4, 0.5, 0.6) para encontrar el balance entre precisi√≥n y recall
4. Cuanto m√°s bajo el umbral, m√°s agresivo prediciendo churn (mayor recall)

**Ventaja**: Simple y efectivo para ajustar la sensibilidad.
**Desventaja**: Requiere experimentaci√≥n para encontrar el umbral √≥ptimo.

| T√©cnica | Ventajas | Desventajas | Mejor Para |
|:---|:---|:---|:---|
| **Original** | Simple, r√°pido | No detecta churn, sesgado | Baseline/Referencia |
| **class_weight='balanced'** | Una l√≠nea de c√≥digo, muy f√°cil | Puede no ser suficiente en desbalanceos extremos | Primer intento, caso general |

| **Undersampling** | Entrena m√°s r√°pido | Pierde informaci√≥n valiosa | Datasets muy grandes |
| **Threshold Adjustment** | Muy flexible, control fino | Requiere tuning y validaci√≥n cruzada | Fine-tuning de modelos existentes |
| **Combinada** | Combina lo mejor de dos mundos | M√°s par√°metros a ajustar | Cuando necesitas m√°xima efectividad |

#### üéØ Recomendaciones Pr√°cticas

**Para Principiantes:**
1. Comienza con `class_weight='balanced'` (T√©cnica 1)
2. Si necesitas mejor rendimiento, prueba con Oversampling (T√©cnica 2)
3. Cambia la m√©trica de evaluaci√≥n a **F1-Score** o **ROC-AUC** en lugar de Accuracy

**Para profesionales:**
1. Combina `class_weight='balanced'` + ajuste de umbral (T√©cnica 5)
2. Usa **Cross-Validation** para validar estabilidad
3. Mide el impacto en negocio (costes de falsos positivos vs falsos negativos)

**Nunca olvides:**
- Siempre eval√∫a en datos de **test separados** 
- No ajustes hiperpar√°metros mirando el test set
- Usa **m√©tricas m√∫ltiples**, nunca solo una
- En datos desbalanceados: **Accuracy es enga√±oso**, usa Recall/F1/AUC

## üìà 5. Curva ROC y AUC <a id='5-roc'></a>

La **Curva ROC (Receiver Operating Characteristic)** es una de las herramientas de evaluaci√≥n m√°s importantes para clasificadores binarios.

### 5.1 Visualizando el Rendimiento de un Clasificador <a id='5.1-visualizacion'></a>


La curva ROC visualiza dos m√©tricas:
- **Tasa de Verdaderos Positivos (TPR)** en el eje Y. Es lo mismo que el **Recall**.
- **Tasa de Falsos Positivos (FPR)** en el eje X. Es la proporci√≥n de negativos reales que fueron incorrectamente clasificados como positivos ( `FP / (FP + TN)` ).

Un buen modelo es aquel que maximiza la TPR mientras minimiza la FPR. Esto se traduce en una curva que se acerca lo m√°s posible a la **esquina superior izquierda** del gr√°fico.

**AUC (Area Under the Curve)**: Es el √°rea bajo la curva ROC. 
- **AUC = 1**: Clasificador perfecto.
- **AUC = 0.5**: Clasificador in√∫til (azar, como lanzar una moneda).
- **AUC < 0.5**: Clasificador peor que el azar.

El AUC es muy √∫til porque resume el rendimiento del modelo en un solo n√∫mero, independientemente del umbral de clasificaci√≥n elegido.

````python
from sklearn.metrics import roc_curve, roc_auc_score

# Necesitamos las probabilidades de la clase positiva, no solo las predicciones
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcular el AUC
auc = roc_auc_score(y_test, y_pred_proba)

print(f"AUC: {auc:.2f}")

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Azar')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
print("Curva ROC graficada.")
print("Este gr√°fico muestra la capacidad del modelo para distinguir entre las clases positivas y negativas a diferentes umbrales de decisi√≥n.")
````

Aunque profundizaremos mucho en este modelo en el pr√≥ximo m√≥dulo, aqu√≠ te damos un adelanto conceptual.

### ü§î ¬øPor qu√© se llama "Regresi√≥n" si estamos clasificando?

Es una buena pregunta. El nombre es hist√≥rico y puede ser confuso. Aunque su nombre contenga la palabra "regresi√≥n", la **Regresi√≥n Log√≠stica es en realidad un modelo de CLASIFICACI√ìN**, no de regresi√≥n.

### üßÆ ¬øC√≥mo funciona conceptualmente?

**Regresi√≥n Log√≠stica** es, en esencia:

1. **Regresi√≥n Lineal + Funci√≥n Especial**: Toma una regresi√≥n lineal (como la que vimos en el tema anterior) y aplica una funci√≥n especial llamada **funci√≥n log√≠stica** o **sigmoid**.

2. **De n√∫meros continuos a probabilidades**: 
   - Una regresi√≥n lineal te da un n√∫mero cualquiera (puede ser -100, +1000, etc.)
   - La funci√≥n log√≠stica transforma ese n√∫mero en una **probabilidad** (un n√∫mero entre 0 y 1)
   
3. **De probabilidades a clases**:
   - Si la probabilidad es > 0.5 ‚Üí Predice clase **1** (S√≠, Churn)
   - Si la probabilidad es ‚â§ 0.5 ‚Üí Predice clase **0** (No, No Churn)
