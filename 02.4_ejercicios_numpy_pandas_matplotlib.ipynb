### Ejercicio 1.2 ‚Äî Operaciones con arrays

Crea dos arrays: `a = [1, 2, 3, 4, 5]` y `b = [10, 20, 30, 40, 50]`

Realiza las siguientes operaciones:
- Suma ambos arrays
- Multiplica ambos arrays elemento por elemento
- Calcula la media de `a`
- Encuentra el valor m√°ximo de `b`

````python
# Tu c√≥digo aqu√≠
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

suma = a + b
producto = a * b
media_a = np.mean(a) # Tambi√©n se puede media_a = a.mean()
max_b = np.max(b)    # Tambi√©n se puede max_b = b.max()

print('Suma:', suma)
print('Producto:', producto)
print('Media de a:', media_a)
print('M√°ximo de b:', max_b)
````

### Ejercicio 1.3 ‚Äî Indexaci√≥n y slicing

Dado el array `arr = np.arange(0, 20)`, obt√©n:
- Los elementos del √≠ndice 5 al 10 (inclusive)
- Los elementos pares
- El array invertido
- Los √∫ltimos 5 elementos

````python
# Tu c√≥digo aqu√≠
arr = np.arange(0, 20)

rango_5_10 = arr[5:11]
pares = arr[arr % 2 == 0]
invertido = arr[::-1]
ultimos_5 = arr[-5:]

print('Array original:', arr)
print('√çndices 5 a 10:', rango_5_10)
print('Pares:', pares)
print('Invertido:', invertido)
print('√öltimos 5:', ultimos_5)
````

### Ejercicio 1.4 ‚Äî Arrays 2D

Crea una matriz 4x4 con n√∫meros del 1 al 16, luego:
- Obt√©n la primera fila
- Obt√©n la √∫ltima columna
- Obt√©n el elemento en la posici√≥n [2, 3] (fila 2, columna 3)
- Calcula la suma de cada columna

````python
# Tu c√≥digo aqu√≠
matriz = np.arange(1,17).reshape(4,4)

primera_fila = matriz[0,:]
ultima_columna = matriz[:,-1]
elemento_2_3 = matriz[2,3]
suma_columnas = matriz.sum(axis=0)

print('Matriz:\n', matriz)
print('Primera fila:', primera_fila)
print('√öltima columna:', ultima_columna)
print('Elemento [2,3]:', elemento_2_3)
print('Suma por columna:', suma_columnas)
````

### Ejercicio 2.5 ‚Äî Agrupaci√≥n

Analiza ventas por producto usando `groupby`.

````python
# Tu c√≥digo aqu√≠
ventas = pd.DataFrame({
    'Producto': ['A', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Cantidad': [10, 15, 8, 20, 12, 5, 18],
    'Precio': [100, 50, 100, 75, 50, 100, 75]
})

# Calcula las ventas totales por producto (cantidad)
ventas_por_producto = ventas.groupby("Producto")["Cantidad"].sum()

# Encuentra el producto m√°s vendido
mas_vendido = ventas_por_producto.max()
mas_vendido = ventas_por_producto.idxmax()

# Calcula el ingreso total por producto
ventas['Ingresos'] = ventas['Cantidad'] * ventas['Precio']
ingreso_por_producto = ventas.groupby("Productos")["Ingresos

print('Ventas por producto:\n', ventas_por_producto)
print('\nProducto m√°s vendido:', mas_vendido)
print('\nIngreso por producto:\n', ingreso_por_producto)
````

### Ejercicio 3.5 ‚Äî Subplots

Crea una figura con 4 subplots (2x2):
1. Gr√°fico de l√≠nea (seno)
2. Gr√°fico de barras
3. Histograma
4. Scatter plot

````python
# Tu c√≥digo aqu√≠
fig, axes = plt.subplots(2, 2, figsize=(5, 5))
# Subplot 1: L√≠nea (seno)
x = np.linspace(0, 2*np.pi, 100)
axes[0, 0].plot(x, np.sin(x), color='blue')
axes[0, 0].set_title('Gr√°fico de Seno')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')

# Subplot 2: Barras
categorias = ['A', 'B', 'C', 'D']
valores = [10, 15, 7, 12]
axes[0, 1].bar(categorias, valores, color='orange')
axes[0, 1].set_title('Gr√°fico de Barras')
axes[0, 1].set_xlabel('Categor√≠as')
axes[0, 1].set_ylabel('Valores')

# Subplot 3: Histograma
np.random.seed(43)  # Para reproducibilidad
data = np.random.randn(1000)
axes[1, 0].hist(data, bins=30, color='purple', alpha=0.7)
axes[1, 0].set_title('Histograma de Datos Aleatorios')
axes[1, 0].set_xlabel('Valor')
axes[1, 0].set_ylabel('Frecuencia')

# Subplot 4: Scatter
x = np.random.rand(50)
y = np.random.rand(50)
axes[1, 1].scatter(x, y, color='green')
axes[1, 1].set_title('Gr√°fico de Dispersi√≥n')
axes[1, 1].set_xlabel('Eje X')
axes[1, 1].set_ylabel('Eje Y')

plt.tight_layout()
plt.show()
````

### An√°lisis de ventas mensual

Combina NumPy, Pandas y Matplotlib para realizar un an√°lisis completo de ventas.

````python
# Tu c√≥digo aqu√≠

# 1. Crear datos de ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
producto_a = [120, 135, 148, 160, 155, 170]
producto_b = [80, 95, 88, 102, 110, 105]

# 2. Crear DataFrame con los datos
df_ventas = pd.DataFrame({"Mes": meses, "Producto_a": producto_a, "Producto_b": producto_b })

# 3. Agregar columna de ventas totales
df_ventas["Total"]=df_ventas["Producto_a"] + df_ventas['Producto_b']

# 4. Calcular estad√≠sticas
media_a = df_ventas['Producto_a'].mean()
media_b = df_ventas['Producto_b'].mean()
# mes_max_ventas = df_ventas['Total'].max() no nos proporciona el mes
mes_max_ventas = df_ventas.loc[df_ventas['Total'].idxmax(), 'Mes']
#              = df_ventas.loc[            5              , 'Mes']                       
# Crecimiento porcentual del Producto A
crecimiento_a = (producto_a[-1] - producto_a[0]) / producto_a[0] * 100
# Crecimiento porcentual del Producto B
crecimiento_b = (producto_b[-1] - producto_b[0]) / producto_b[0] * 100
# Imprimir resultados
print('DataFrame de ventas:\n', df_ventas)
print(f'\nMedia Producto A: {media_a:.2f}')
print(f'Media Producto B: {media_b:.2f}')
print(f'Mes con mayores ventas: {mes_max_ventas}')
print(f'Crecimiento Producto A: {crecimiento_a:.2f}%')
print(f'Crecimiento Producto B: {crecimiento_b:.2f}%')
````

````python
# 5. Crear visualizaciones

# a) Gr√°fico de l√≠neas comparando ambos productos
plt.figure(figsize=(10, 6))
plt.plot(df_ventas['Mes'], df_ventas['Producto_a'], marker='X', label='Producto A', color='blue')
plt.plot(df_ventas['Mes'], df_ventas['Producto_b'], marker='o', label='Producto B', color='green')
plt.title('Ventas Mensuales de Productos A y B')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# b) Gr√°fico de barras con ventas totales por mes
plt.figure(figsize=(10, 6))
plt.bar(df_ventas['Mes'], df_ventas['Total'], color='purple', alpha=0.7)
plt.title('Ventas Totales por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.show()
````

````python
# c) Gr√°fico de barras agrupadas (productos lado a lado)
x = np.arange(len(meses))  # Posiciones en el eje x
width = 0.35  # Ancho de las barras 
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, df_ventas['Producto_a'], width, label='Producto A', color='blue')
# x - width/2 para Producto A y x + width/2 para Producto B
plt.bar(x + width/2, df_ventas['Producto_b'], width, label='Producto B', color='green')
plt.xticks(x, meses) # Etiquetas de los meses
plt.title('Ventas Mensuales de Productos A y B (Agrupadas)')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.show()  
````

````python
# 6. Guardar DataFrame en CSV
df_ventas.to_csv("ventas_productos.csv", index=False)
````
